```python
results_df = pd.read_csv('game_results.csv')
```

```python
week_1 = results_df[results_df['week'] == 1]
```

```python
print(week_1.to_string())
```

```python
wl = []
```

```python
for row in week_1.iterrows():
    if row[1]['team1_winner']:
        winner = row[1]['team1_location']
        loser = row[1]['team2_location']
    else:
        winner = row[1]['team2_location']
        loser = row[1]['team1_location']
    wl.append({"winner": winner,"loser": loser})

wl
```

```python
def replace_team(row):
    for w in wl:
        if row['school'] == w['winner']:
            return w['winner']
        elif row['school'] == w['loser']:
            return w['winner']
```

```python
baseline['new_team'] = baseline.apply(replace_team, axis=1)
```
